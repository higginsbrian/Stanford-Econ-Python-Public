### Simulation of AR(1) Process

Consider the following process

$$x_t = \rho_x x_{t - 1} + \epsilon^x_t$$

$$\epsilon_x \sim N(0, \sigma^2)$$

**GOAL** Simulate the Series

* STEP 1: Generate random draws of $\epsilon_t$
* STEP 2: Generate the implied evolution of the serries given the draws

**Excercises**

1. Evaluate estimated standard deviation of series 
2. Construct Figure for evolution of AR(1) process 
3. Do the same for AR(2) process

```python
import numpy as np
from scipy.stats import norm
#import plotly.express as px
import matplotlib.pyplot as plt
```

```python
# Primitives
σ = 2
ρ = 0.95
N_sim = 100 # Number of time periods
```

```python
# Initialize array to save values
x_sim = np.zeros(N_sim + 1)
```

```python
# Set random seed
np.random.seed(seed = 712)

# Define normal distribution with mean 0 and std σ
ϵ_dist = norm(0, σ)

# Draw N_sim + 1 values from the above distribution
ϵ_vec = ϵ_dist.rvs(N_sim)
```

```python
# AR(1) simulation
for i in range(1, N_sim + 1):
    x_sim[i] = ρ*x_sim[i - 1] + ϵ_vec[i - 1]
```

```python
st_dev_est = np.std(ϵ_vec)
print('Estimated standard deviation = ', round(st_dev_est, 3))
```

```python
# Plot results (using plotly)
plt.plot( x_sim)
plt.xlabel('Time Period')
plt.ylabel('X')
plt.title('Simulation: AR(1) Process')
```
